# Global Trend Over Time

This chart will show the total number of animals slaughtered globally for each type of animal over time.

```python
global_trend = df.groupby('Year').sum()

plt.figure(figsize=(10, 6))

for column in global_trend.columns:
    plt.plot(global_trend.index, global_trend[column], label=column)

plt.title('Global Animal Slaughter Trend (1961-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Animals Slaughtered')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# Top Countries by Animal Slaughter (For a Specific Year)

This bar chart will display the top 10 countries with the highest number of a particular animal slaughtered (e.g., chickens) in a given year. You can specify the year and the animal category.

```python
year = 2020
animal = 'Number of chickens slaughtered to produce meat'

year_data = df[df['Year'] == year]

top_countries = year_data[['Entity', animal]].sort_values(by=animal, ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_countries['Entity'], top_countries[animal], color='skyblue')
plt.title(f'Top 10 Countries by {animal} Slaughtered in {year}')
plt.xlabel('Number of Animals Slaughtered')
plt.ylabel('Country')
plt.gca().invert_yaxis()  
plt.show()
```

# Stacked Bar Chart: Animal Distribution by Country (For a Specific Year)

This chart will show the slaughter numbers of different animal categories in several countries for a specific year. You can modify the countries and year accordingly.

```python
year = 2020
countries = ['Afghanistan', 'India', 'Brazil', 'United States', 'China']

filtered_data = df[(df['Year'] == year) & (df['Entity'].isin(countries))]

animal_columns = ['Number of chickens slaughtered to produce meat', 
                  'Number of ducks slaughtered to produce meat',
                  'Number of pigs slaughtered to produce meat',
                  'Number of sheep slaughtered to produce meat',
                  'Number of cattle slaughtered to produce meat']

filtered_data.set_index('Entity', inplace=True)

filtered_data[animal_columns].plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title(f'Animal Slaughter Distribution by Country in {year}')
plt.xlabel('Country')
plt.ylabel('Number of Animals Slaughtered')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1)) 
plt.show()
```

# Global Stacked Area Chart

The chart displays the total number of animals slaughtered globally for different categories, stacking them over time. It visually shows how the proportion of each animal type contributes to the overall slaughter numbers across years.


```python
global_trend = df.groupby('Year').sum()

animal_columns = ['Number of chickens slaughtered to produce meat', # Select the relevant columns for animals
                  'Number of ducks slaughtered to produce meat',
                  'Number of pigs slaughtered to produce meat',
                  'Number of sheep slaughtered to produce meat',
                  'Number of cattle slaughtered to produce meat']

plt.figure(figsize=(10, 6))
plt.stackplot(global_trend.index, global_trend[animal_columns].T, labels=animal_columns, alpha=0.8)

plt.title('Global Animal Slaughter Proportions Over Time (1961-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Animals Slaughtered')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# Global Correlation Heatmap

This heatmap shows how different animal categories are correlated globally. A correlation value of +1 means a perfect positive correlation (e.g., both types of animals are slaughtered more as time progresses), whereas -1 means a negative correlation (when one increases, the other decreases).

```python
global_trend = df.groupby('Year').sum()

animal_columns = ['Number of chickens slaughtered to produce meat', 
                  'Number of ducks slaughtered to produce meat',
                  'Number of pigs slaughtered to produce meat',
                  'Number of sheep slaughtered to produce meat',
                  'Number of cattle slaughtered to produce meat']

correlation_matrix = global_trend[animal_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

plt.title('Correlation Heatmap of Animals Slaughtered Globally (1961-2022)')
plt.show()
```

# Country-Specific Correlation Heatmap

This heatmap focuses on the correlation between animal categories in a single country. It can help identify if certain animals are slaughtered together in larger quantities or if the numbers are inversely related.

```python
country = 'Australia' # Filter data for a specific country
country_data = df[df['Entity'] == country]

correlation_matrix_country = country_data[animal_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_country, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)

plt.title(f'Correlation Heatmap of Animals Slaughtered in {country} (1961-2022)')
plt.show()
```

# Global Pie Chart

This pie chart shows how much each type of animal contributes to the total number of animals slaughtered globally in the selected year.

```python
year = 2022  # Select the year for comparison

year_data = df[df['Year'] == year].sum()

animal_columns = ['Number of chickens slaughtered to produce meat', 
                  'Number of ducks slaughtered to produce meat',
                  'Number of pigs slaughtered to produce meat',
                  'Number of sheep slaughtered to produce meat',
                  'Number of cattle slaughtered to produce meat']

animal_slaughter_data = year_data[animal_columns]

plt.figure(figsize=(8, 8))
plt.pie(animal_slaughter_data, labels=animal_columns, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title(f'Proportion of Animals Slaughtered Globally in {year}')
plt.axis('equal') 
plt.show()
```

# Country-Specific Pie Chart

This pie chart breaks down the proportions of different animals slaughtered within a specific country for the selected year.

```python
# Define the country and year
country = 'Africa'
year = 2020

country_data = df[(df['Entity'] == country) & (df['Year'] == year)]

if not country_data.empty:
    country_data = country_data.iloc[0]
    
    
    animal_columns = [col for col in animal_columns if col in df.columns]

    country_animal_slaughter_data = country_data[animal_columns]

    country_animal_slaughter_data = country_animal_slaughter_data.dropna()

    if country_animal_slaughter_data.empty:
        print("All data is NaN.")
    else:
        plt.figure(figsize=(8, 8))
        plt.pie(
            country_animal_slaughter_data,
            labels=country_animal_slaughter_data.index,
            autopct='%1.1f%%',
            startangle=140,
            colors=sns.color_palette('Set3')
        )
        plt.title(f'Proportion of Animals Slaughtered in {country} in {year}')
        plt.axis('equal') 
        plt.show()
else:
    print(f"No data found for {country} in {year}")
```

# Histogram

This code will generate a line plot with different lines for each type of animal. If you specifically want a histogram, you might need to adjust your approach to fit your data, or consider using a stacked bar chart or area plot for a more suitable visualization.

```python
animal_columns = ['Number of chickens slaughtered to produce meat',
                  'Number of ducks slaughtered to produce meat',
                  'Number of pigs slaughtered to produce meat',
                  'Number of sheep slaughtered to produce meat',
                  'Number of cattle slaughtered to produce meat']

all_slaughter_numbers = df[animal_columns].values.flatten()

plt.figure(figsize=(12, 8))
plt.hist(all_slaughter_numbers, bins=30, alpha=0.75, color='blue')
plt.title('Distribution of Animal Slaughter Numbers (1961-2022)')
plt.xlabel('Number of Animals Slaughtered')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Scattter plot

Scatter plots showing the relationship between different types of animals slaughtered in the same country, for instance, comparing the number of chickens versus cattle. creates a scatter plot for each pair of columns. Adjust the alpha value to control the transparency of the points.

```python
animal_columns = {
    'Chickens vs Cattle': ('Number of chickens slaughtered to produce meat', 
                           'Number of cattle slaughtered to produce meat'),
}

plt.figure(figsize=(12, 8))

for title, (x_col, y_col) in animal_columns.items():
    plt.scatter(df[x_col], df[y_col], alpha=0.5, label=title)

plt.title('Scatter Plot of Animal Slaughter Numbers')
plt.xlabel('Number of Chickens Slaughtered')
plt.ylabel('Number of Cattle Slaughtered')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
